```python
def brands_encoding(data):
    """Function takes data.CarName and returns encoded column."""
    brands = {'alfa-romero': 'alfa-romero',
              'audi': 'audi',
              'bmw': 'bmw',
              'chevrolet': 'chevrolet',
              'dodge': 'dodge',
              'honda': 'honda',
              'isuzu': 'isuzu',
              'jaguar': 'jaguar',
              'maxda': 'mazda',
              'mazda': 'mazda',
              'buick': 'buick',
              'mercury': 'mercury',
              'mitsubishi': 'mitsubishi',
              'nissan': 'nissan',
              'peugeot': 'peugeot',
              'plymouth': 'plymouth',
              'porsche': 'porsche',
              'porcshce': 'porsche',
              'renault': 'renault',
              'saab': 'saab',
              'subaru': 'subaru',
              'toyota': 'toyota',
              'toyouta': 'toyota',
              'vokswagen': 'vw',
              'volkswagen': 'vw',
              'vw': 'vw',
              'volvo': 'volvo'}
    return data.map(lambda x: brands[x.split()[0].lower()])
```

```python
def categorical_encoding(data):
    """Function takes dataframe with categorical columns and encodes them from zero to length."""
    
    symboling_dummies = pd.get_dummies(data['symboling'], drop_first=False, prefix='symboling')

    data = pd.concat([data, symboling_dummies], axis=1)

    cars_categorical = data.select_dtypes(include=['object'])

    cars_dummies = pd.get_dummies(cars_categorical, drop_first=False)
    
    data = data.drop(list(cars_categorical.columns), axis=1)
    
    data = pd.concat([data, cars_dummies], axis=1)
        
    return data
```

```python
def numerical_transformer(data):
    numerical = ['wheelbase','carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize',
                 'boreratio', 'stroke','compressionratio', 'horsepower', 'peakrpm', 'citympg',
                 'highwaympg']
        
    num_transformer = Pipeline([
        ('imputer', SimpleImputer()),
        ('scaler', MinMaxScaler())
    ])

    num_transformer.fit(data[numerical])
    data[numerical] = num_transformer.transform(data[numerical])
    
    return data
```

```python
def pipe(data):
    
    data.drop_duplicates()
    
    data['brands'] = brands_encoding(data.CarName)
    
    data = categorical_encoding(data.drop(columns=['car_ID', 'CarName']))
    
    data = numerical_transformer(data.drop(columns='symboling'))
    
    return pd.DataFrame(data)
```

```python
data['brands'] = brands_encoding(data.CarName)
data
```

```python
categorical_encoding(data.drop(columns=['car_ID', 'CarName']))
```

```python
X = data.drop(columns='price')
y = data[['price']]
```

```python
X = pipe(X)
```

```python
model = LinearRegression()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    train_size=0.7,
                                                    test_size = 0.3, random_state=100)
```

```python
model.fit(X_train, y_train)
```

```python
# predict 
y_pred = model.predict(X_test)

# metrics
from sklearn.metrics import r2_score

print(r2_score(y_true=y_test, y_pred=y_pred))
```
